- The number of trials (n) is 10 (the sample size).

Let's calculate each probability:

   This is the probability of 10 consecutive successes (non-defective items).

2.D. Probability that three or more of the items are defective:

Let's calculate each probability:

## 3.A. Probability that in a given week he will sell some cars (at least 1 car):
   This is the complement of the probability that he sells zero cars.

   P(X ≥ 1) = 1 - P(X = 0)

   Using the Poisson distribution formula:
   P(X = 0) = (e^(-λ) * λ^0) / 0! = e^(-3) ≈ 0.0498

   Therefore, P(X ≥ 1) ≈ 1 - 0.0498 ≈ 0.9502

## 3.B. Probability that in a given week he will sell 2 or more but less than 5 cars (2 ≤ X < 5):
   This is the sum of the probabilities for X = 2, 3, and 4.

   P(2 ≤ X < 5) = P(X = 2) + P(X = 3) + P(X = 4)

   Using the Poisson distribution formula for each value of X:
   P(X = 2) = (e^(-λ) * λ^2) / 2! = (e^(-3) * 3^2) / 2! ≈ 0.2240
   P(X = 3) = (e^(-λ) * λ^3) / 3! = (e^(-3) * 3^3) / 3! ≈ 0.2240
   P(X = 4) = (e^(-λ) * λ^4) / 4! = (e^(-3) * 3^4) / 4! ≈ 0.1680

   Therefore, P(2 ≤ X < 5) ≈ 0.2240 + 0.2240 + 0.1680 ≈ 0.6160



## 3.C. To plot the Poisson distribution function for cumulative probability of cars sold per-week vs. the number of cars sold per week as shown below : 


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Average rate
lambda_ = 3

# Number of cars sold per week (0 to 10)
x = np.arange(0, 11)

# Poisson probability mass function
pmf = poisson.pmf(x, lambda_)

# Cumulative probability (CDF)
cdf = np.cumsum(pmf)

# Create the plot
plt.figure(figsize=(8, 4))
plt.plot(x, cdf, marker='o', linestyle='-')
plt.title('Poisson Distribution for Car Sales per Week')
plt.xlabel('Number of Cars Sold per Week')
plt.ylabel('Cumulative Probability')
plt.grid(True)
plt.show()


#This code will generate a plot that shows the cumulative probability of cars sold per week for the given Poisson distribution with an average rate of 3 cars sold per week.
```

In hypothesis testing, we typically define the null hypothesis (H0) and alternative hypothesis (H1) based on the question we are trying to answer. In this case, we are investigating if there is a significant difference in funds raised between companies that are still operating and companies that have closed down. Here are the null and alternative hypotheses:

**Null Hypothesis (H0):**
There is no significant difference in the mean funds raised between companies that are still operating and companies that have closed down.
μ operating = μ closed


**Alternative Hypothesis (H1):**
There is a significant difference in the mean funds raised between companies that are still operating and companies that have closed down.
 μ operating not equal to  μ closed

Here, μ operating  represents the mean funds raised by companies that are still operating, and μ closed represents the mean funds raised by companies that have closed down. The alternative hypothesis (H1) states that the means are not equal, indicating a significant difference.
